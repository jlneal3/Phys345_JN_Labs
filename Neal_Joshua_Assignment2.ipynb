```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
point = np.loadtxt('spring_data.csv', delimiter=',', skiprows=1)
for i in range(14):           #This iterates through each component in the array created by the np.loadtxt file  [NOT NEEDED IN FUTURE]
    plt.scatter(point[i][0],point[i][1], color = "red")
plt.scatter(point[14][0], point[14][1], color = "red", label = "data") # this is just the final data point, so I could add the legend without it repeating
m =  np.linspace(0,1.5,15)

def spring_model(k,m_dum):
    x = (9.81/k)*m_dum           
    return x

test = np.array([10,20,25,30,40])
for values in test:                # Iterates through the guess values of the array
    x_dis = spring_model(values,m)
    plt.plot(m,x_dis, label = values)
    Res = x_dis- point[:,1] #iterates only through the displacement values of the data and subtracts it from the model
    MSE = np.mean((spring_model(25,m)-point[:,1])**2)  #Only uses the k=25 array and turns it into the mean sum sqaure
    print("Residuals for k = ", values, Res)

plt.legend(loc = "upper left")
print( "MSE for k = 25 ", MSE)
plt.xlabel('Mass')
plt.ylabel('Displacement')
# When k = 25, you get a line that best fits the data
#Increasing k leads to the model lines becoming flatter
# As you move K away from the best value, it increases the MSE. For example, at 25, the MSE = 0.0008, at 20 =K, the MSE = 0.006 and at 30 = k MSE = 0.005

#If you increase k, does the slope of your model line become steeper or flatter?
#How does the MSE change if you move k away from your "best" value? 
```
