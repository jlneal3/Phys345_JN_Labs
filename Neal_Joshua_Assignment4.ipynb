```python
import numpy as np
import matplotlib.pyplot as plt

dt = [10.0,0.1,1] #mins
for i in range (3):
    dt[i]
    N = 1000
    lamb = 0.05 #mins^-1
    t_tot = 100 #mins
    t = 1
    
    N_list = []
    t_list = []
    
    while t<t_tot:   
        delta_N = -lamb*N*dt[i] # defining the change/ incrementing
        t+=dt[i]
        N = N + delta_N # defining the final population
        N_list.append(N)
        t_list.append(t)
    
    label = ["dt = 10.0","dt= 0.1","dt=1.0"]
    color = ['green','blue','yellow']
    
    plt.plot(t_list, N_list,label = label[i], color = color[i])
    plt.xlabel('Time')
    plt.ylabel('Nuclei') 


#Analytical solution
t_2 =  np.linspace(0,100,25)

def decay_func(lamb_2,t_dum):
    N = 1000*np.e**(-lamb_2*t_dum)
    return N
nuclei = decay_func(t_2,0.05)
plt.scatter(t_2,nuclei, label = 'Model B')
plt.legend()
plt.show()

# The sim
```
